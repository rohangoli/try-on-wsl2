### Bag of Words - NLP Technique

```python
dataset['Score'].unique()
```

```python
# Filter Neutral Values like 3
dataset_nonet = dataset[dataset['Score']!=3]
```

```python
dataset_nonet.head()
```

```python
X=dataset_nonet['Text']
```

```python
# Replace Integer Scores with 0/1 for better classification
y_dict = {1:0,2:0,4:1,5:1}
y = dataset_nonet['Score'].map(y_dict)
```

```python
## Count Vectorizer for Bag of Words

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
## Create Bag of Words Matrix
X_cv = cv.fit_transform(X)
```

```python
X_cv.shape
```

### Check Model Accuracy

```python
## Split Data for Test and Train

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_cv,y)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr_model = lr.fit(X_train,y_train)
```

```python
lr_model.score(X_test,y_test)
```

### Fetch Top 20 Postive and Negative words each 

```python
w = cv.get_feature_names()
```

```python
coef = lr_model.coef_.tolist()[0]
```

```python
coef_dataset = pd.DataFrame({'Word':w,'Coefficient':coef})
```

```python
coef_dataset
```

```python
coef_words = coef_dataset.sort_values(['Coefficient','Word'],ascending=False)
coef_words
```

```python
## Top 20 Positive Words
coef_words.head(20)
```

```python
## Top 20 Negative words
coef_words.tail(20)
```

### Automate NLP Model & ML Model 

```python
def text_fit(X,y,nlp_model,ml_model):
    X_cv = nlp_model.fit_transform(X)
    print('features:{}'.format(X_cv.shape[1]))
    X_train, X_text, y_train, y_test = train_test_split(X_cv,y)
    ml = ml_model.fit(X_train,y_train)
    acc = ml.score(X_test,y_test)
    print(acc)
    
    words =  nlp_model.get_feature_names()
    coef = ml.coef_.tolist()[0]
    coef_dataset = pd.DataFrame({'Word':words,'Coefficient':coef})
    coef_words = coef_dataset.sort_values(['Coefficient','Word'],ascending=False)
    print('\n')
    print("Top 20 Positive words:")
    coef_words.head(20)
    print('\n')
    print("Top 20 Negative words:")
    coef_words.tail(20)
    
    
```
